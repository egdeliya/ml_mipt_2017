```python
# считывает данные
def read_file(file_name, item):
    ts = pd.read_csv(file_name, sep=";", decimal='.', parse_dates=True, dayfirst=True, index_col='Timestamp')
    ts.sort_index(inplace=True)
    ts = ts.iloc[:, item]
    ts.fillna(0, inplace=True)
    return ts
```

```python
# для анализа данных, строит декомпозицию и выводит значение критерия Дики-Фуллера
def analyse_data(ts, resampling_per="D"):
    res = sm.tsa.seasonal_decompose(ts.resample(resampling_per).sum())
    resplot = res.plot()
    test = sm.tsa.adfuller(ts.get_values())
    print('Критерий Дики-Фуллера\np-value: {}'.format(test[1]))
    print("table distribution {}".format(test[4]))
```

```python
from scipy import stats
from copy import deepcopy, copy
```

##### Построим графики ACF/PACF для определения параметров $p$ и $q$

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(diff_train_log.values.squeeze(), lags=40, ax=ax)
ax = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(diff_train_log, lags=40, ax=ax)
```

* #### Из графика ACF можно предположить возможные значения для $q$ - 1
* #### Из графика PACF можно предположить возможные значения для $p$ - 1-7

#### Переберём возможные параметры ($p$, $d$, $q$)

```python
p = list(range(5))
q = [0, 1]
d = [0, 1]
```

```python
parameters = [(_p, _d, _q) for _p in p
                              for _d in d
                                  for _q in q]
```

```python
len(parameters)
```

#### Выберем лучшую модель, используя критерий AIC. Здесь нужно учесть, что для некоторых параметров модель ARIMA не обучается

```python
import warnings
from tqdm import tqdm
```

```python
%%time
scores = []
aic = []
warnings.filterwarnings('ignore')

for params in tqdm(parameters):
    try:
        model = sm.tsa.ARIMA(diff_train_log, params).fit()
    except:
        print("wrong parameters {}".format(params))
        continue
    res = model.aic
    aic.append(res)
    scores.append([params, res])
```

```python
res_table = pd.DataFrame(scores)
res_table.columns = ["parameters", "aic"]
res_table.sort_values(by="aic", ascending=True).head()
```

```python
best_params = res_table.sort_values(by="aic", ascending=True)['parameters'][5]
print(best_params)
# best_params = (2, 0, 1)
second_params = res_table.sort_values(by="aic", ascending=True)['parameters'][8]
```

```python
best_model = sm.tsa.ARIMA(diff_train_log, best_params).fit()
```

#### Проведём анализ остатков нашей модели

```python
resid = best_model.resid
```

#### Построим график

```python
plt.subplot(211)
resid.plot(figsize=(15, 7))
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=48, ax=ax)

print("Анализ остатков")
```

#### По графику можно сказать, что остатки стационарны не скорреллированы. Проведём также тест на корреляцию Льюнга-Бокса

```python
# Ljung-Box test 
pd.DataFrame({'lags':range(30), 'pvalue':sm.stats.diagnostic.acorr_ljungbox(resid.values, lags=30)[1], 
              'critial':np.array([0.05]*30)}).set_index('lags').plot(figsize=(15,5))
```

#### Тесть подтверждает, что остатки нескоррелированы. Проверим на несмещённость с помощью критерия Стьюдента и на стационарность с помощью критерия Дики-Фуллера

```python
print("Критерий Стьюдента: p={}\n".format(stats.ttest_1samp(resid, 0)[1])) 

test_full = sm.tsa.stattools.adfuller(resid)
print("Критерий Дики-Фуллера: p={} \ntable distribution: {}".format(test_full[1], test_full[4]))
```
